```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Step 1: Set up Chrome in headless mode
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=options)

# Step 2: Open NTES
driver.get("https://enquiry.indianrail.gov.in/mntes/")

# Step 3: Search for Train
search_box = driver.find_element(By.ID, "trainInput")  # Update with correct ID
search_box.send_keys("00132")
time.sleep(1)  # Wait for suggestions to load
search_box.send_keys("\n")  # Press Enter

time.sleep(3)  # Wait for results to load

# Extract train heading
train_heading = driver.find_element(By.CLASS_NAME, "train-heading").text
print("Train Info:", train_heading)

# Extract all station rows
stations = []
rows = driver.find_elements(By.CLASS_NAME, "station-row")
for row in rows:
    name = row.find_element(By.CLASS_NAME, "station-name").text
    sched_arr = row.find_element(By.CLASS_NAME, "sched-arrival").text
    act_arr = row.find_element(By.CLASS_NAME, "act-arrival").text
    stations.append({
        "station": name,
        "scheduled_arrival": sched_arr,
        "actual_arrival": act_arr
    })

print(stations)

driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

# Train details
TRAIN_NUMBER = "00132"   # Example: 00132 TDN-REWA BHARAT GAURAV SP
JOURNEY_DATE = "26-08-2025"  # Format: DD-MM-YYYY

# NTES URLs
BASE_URL = "https://enquiry.indianrail.gov.in/mntes/"
RUNNING_STATUS_URL = "https://enquiry.indianrail.gov.in/mntes/tr?opt=TrainRunning&subOpt=FindRunningInstance"

# Start session to maintain cookies
session = requests.Session()
session.headers.update({
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36"
})

# Step 1: Initial GET to load cookies
print("[*] Getting initial session...")
initial_resp = session.get(BASE_URL)
if initial_resp.status_code == 200:
    print("[+] Session initialized.")
else:
    print("[-] Failed to get initial session.")
    exit()

# Step 2: POST request to find running instance
payload = {
    "trainNo": TRAIN_NUMBER,
    "jDate": JOURNEY_DATE
}

print("[*] Fetching train running status...")
resp = session.post(RUNNING_STATUS_URL, data=payload)

if resp.status_code != 200:
    print(f"[-] Failed to fetch data. Status Code: {resp.status_code}")
    exit()

# Step 3: Parse response HTML
soup = BeautifulSoup(resp.text, "html.parser")

# Extract train details
train_name = soup.find("span", {"id": "trainName"}).get_text(strip=True) if soup.find("span", {"id": "trainName"}) else "N/A"
print(f"\nTrain: {TRAIN_NUMBER} - {train_name}\n")

# Extract stations info
stations = soup.select("div.runningStatus div.route")  # Adjust selector based on HTML structure

if not stations:
    print("[-] No station data found.")
else:
    print("[+] Train Running Details:\n")
    for st in stations:
        st_name = st.find("div", class_="station").get_text(strip=True) if st.find("div", class_="station") else "Unknown"
        sched_arr = st.find("span", class_="schArr").get_text(strip=True) if st.find("span", class_="schArr") else "-"
        sched_dep = st.find("span", class_="schDep").get_text(strip=True) if st.find("span", class_="schDep") else "-"
        act_arr = st.find("span", class_="actArr").get_text(strip=True) if st.find("span", class_="actArr") else "-"
        act_dep = st.find("span", class_="actDep").get_text(strip=True) if st.find("span", class_="actDep") else "-"
        status = st.find("span", class_="delay").get_text(strip=True) if st.find("span", class_="delay") else "On Time"

        print(f"Station: {st_name}")
        print(f"  Scheduled Arrival: {sched_arr}, Scheduled Departure: {sched_dep}")
        print(f"  Actual Arrival: {act_arr}, Actual Departure: {act_dep}")
        print(f"  Status: {status}\n")
```
