```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test_data = pd.read_csv("test.csv")
test_data
```

```python
train_data = pd.read_csv("train.csv")
train_data.head(25)
```

```python
sample = pd.read_csv("sample_submission.csv")
sample = pd.merge(sample, test_data, on = "PassengerId")
sample
```

```python
train_data.describe()
```

```python
train_data.isnull().any()
```

```python
dropped_data = train_data.dropna()
dropped_data
```

```python
dropped_data.describe()
```

```python
print(dropped_data['HomePlanet'].unique())
print(dropped_data['Destination'].unique())
```

```python
dropped_data = pd.get_dummies(dropped_data, columns = ['HomePlanet'])
dropped_data = pd.get_dummies(dropped_data, columns = ['Destination'])
```

```python
dropped_data.columns
```

```python
from sklearn.tree import DecisionTreeClassifier

y = dropped_data['Transported']

X = dropped_data[['CryoSleep', 'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'HomePlanet_Earth', 'HomePlanet_Europa', 'HomePlanet_Mars','Destination_55 Cancri e', 'Destination_PSO J318.5-22','Destination_TRAPPIST-1e']]
```

```python
X.head()
```

```python
iowa_model = DecisionTreeClassifier(random_state=0)
iowa_model.fit(X,y)
```

```python
predictions = iowa_model.predict(X)
predictions
```

```python
y.head()
```

```python
test_data = pd.get_dummies(test_data, columns = ['HomePlanet'])
test_data = pd.get_dummies(test_data, columns = ['Destination'])
```

```python
test_data = test_data.drop(columns=['Cabin', 'Name', 'PassengerId'])  # keep only features
```

```python
original_predictions = iowa_model.predict(test_data)
print(original_predictions.size)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y, predictions)
acc
```

```python
test_acc = accuracy_score(sample['Transported'], original_predictions)
test_acc
```
