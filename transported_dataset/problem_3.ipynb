```python
age_data = train_data.sort_values("Age")
px.histogram(age_data["Age"])
```

```python
fig = make_subplots(rows = 3, cols = 1)
p_1 = px.scatter(data_frame=age_data, x = "Age", y = "CryoSleep")
p_2 = px.scatter(data_frame=age_data, x = "Age", y = "RoomService")
p_3 = px.bar(data_frame=age_data, x = "Age", y = "Transported") #age doesn't seem to be significant (might as well drop it)
fig.add_trace(p_1.data[0], row = 1, col = 1)
fig.add_trace(p_2.data[0], row = 2, col = 1)
fig.add_trace(p_3.data[0], row = 3, col = 1)
fig.show()
```

```python
fig = make_subplots(rows = 3, cols = 1)
p_1 = px.scatter(data_frame=age_data, x = "VIP", y = "CryoSleep")
p_2 = px.scatter(data_frame=age_data, x = "VIP", y = "RoomService")
p_3 = px.bar(data_frame=age_data, x = "VIP", y = "Transported") #age doesn't seem to be significant (might as well drop it)
fig.add_trace(p_1.data[0], row = 1, col = 1)
fig.add_trace(p_2.data[0], row = 2, col = 1)
fig.add_trace(p_3.data[0], row = 3, col = 1)
fig.show()
```
